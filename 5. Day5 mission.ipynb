```python
import numpy as np

import pandas as pd

d=pd.read_csv("./netflix_titles.csv")
s_d=d['country'].value_counts(sort=True)
max_num=s_d[0]
g_d=d['country'].groupby(by=d['country']).count()
d_d=dict(g_d)
for i in d_d.keys():
    if ',' not in i:
        if d_d[i]==max_num:
            print(i,max_num)
            break
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

bit=pd.read_csv("./BitCoin.csv")
start=pd.to_datetime('2016-05-28')
end=pd.to_datetime('2017-06-30')
bit['Date']=pd.to_datetime(bit['Date'])

mask=((bit['Date']>=start) & (bit['Date']<=end))
mask_bit=bit[mask]

s_bit=mask_bit.sort_values(by=['Date'])
r_bit=s_bit['Open'].rolling(window=5).mean()
r_bit=r_bit.dropna()
s_bit=s_bit[s_bit['Date']>=pd.to_datetime("2016-06-01")]


eth=pd.read_csv("./ETH_day.csv")

eth['Date']=pd.to_datetime(eth['Date'])
mask=((eth['Date']>=start) & (eth['Date']<=end))
mask_eth=eth[mask]



s_eth=mask_eth.sort_values(by=['Date'])
r_eth=s_eth['Open'].rolling(window=5).mean()
r_eth=r_eth.dropna()
s_eth=s_eth[s_eth['Date']>=pd.to_datetime("2016-06-01")]





plt.xlabel("Datetime")
plt.ylabel("Open")

plt.plot(s_bit['Date'],r_bit,color='#f2a900',label="Bitcoin")

plt.plot(s_eth['Date'],r_eth,color='#3c3c3d',label="Etherium")
plt.legend()
plt.figure(figsize=(1000,1000))
plt.show()



```
