### 1. 본문에서 언급된 Feature를 제외하고 유의미한 Feature를 1개 이상 찾아봅시다.

- Hint : Fare? Sibsp? Parch?


- 가설 1 : Fare로는 구분하기 어려울 것이라고 판단, 이유는 1등급과 2등급 가격을 정확하게 알 수 없으며 실제 플롯을 그릴 때 기준점을 설정하기가 어려웠음

- 가설 2: 혼자보다는 누군가와 같이 있을 때 더 살아남기 용이할 것이라고 판단, 적어도 다른 지인의 도움을 받을것 같다는 생각



```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

titanic=pd.read_csv("./titanic/train.csv")

sss=titanic[['SibSp','Survived']].groupby(['SibSp']).mean()

plt.plot(sss['Survived'])
plt.show()


sss=titanic[['Parch','Survived']].groupby(['Parch']).mean()
plt.plot(sss['Survived'])
plt.show()
# s

```

- 두 그래프를 보았을 때 사람이 많으면 생존 확률이 드물지만 1~3까지는 유의미한 생존률을 보여준다고 할 수 있다

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

fifa=pd.read_csv("./fifa/data.csv")

che=fifa['Position']!="GK"

fifa=fifa[che]

plt.scatter(fifa['Dribbling'],fifa['Overall'])
plt.show()

plt.scatter(fifa['BallControl'],fifa['Overall'])
plt.show()


plt.scatter(fifa['Composure'],fifa['Overall'])
plt.show()

```
